```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
import tikzplotlib
```

```python
plt.rcParams["font.family"] = "serif"
plt.rcParams["mathtext.fontset"] = "dejavuserif"
```

```python
f, ax = plt.subplots(1, 1, figsize=(6, 6))

ax.scatter([1, 2, 8 / 5], [2, 1, 4 / 5], color="black")
ax.set_xlim([0, 2.6])
ax.set_xticks(np.arange(0, 2.7, 0.2))

ax.set_xlabel(r"$s_0$")

ax.set_ylim([0, 2.6])
ax.set_ylabel(r"$s_1$")
ax.set_yticks(np.arange(0, 2.7, 0.2))


ax.annotate(r"$v_w$", (2 - 0.13, 1 + 0.01))
ax.annotate(r"$B^\pi v_w$", (1 - 0.25, 2 + 0.01))
ax.annotate(r"$\Pi B^\pi v_w$", (8 / 5 - 0.325, 4 / 5 + 0.01))

ax.arrow(
    0,
    0,
    2.2,
    1.1,
    length_includes_head=True,
    linewidth=0.25,
    color="black",
    head_width=0.03,
    overhang=1,
)
ax.annotate(r"$w$", (2.2, 1.15))

ax.plot([8 / 5, 1], [4 / 5, 2], color="black", linestyle="dashed", linewidth=0.75)
rect = patches.Rectangle(
    (8 / 5, 4 / 5),
    0.1,
    0.1,
    linewidth=1,
    edgecolor="black",
    angle=np.rad2deg(np.arccos(2/np.sqrt(5))),
    facecolor="none",
)
ax.add_patch(rect)

plt.gca().set_aspect("equal")

plt.savefig("../images/7.4.4.pdf", dpi=300, bbox_inches='tight')
plt.tight_layout()
plt.show()
```
